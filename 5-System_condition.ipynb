## Билет 5
### Обусловленность систем линейных алгебраических уравнений. Оценка относительной погрешности.

Вспомним самое важное из линейной алгебры.

**Определение.** *Нормой матрицы $A$* называется величина
$$ \| A \| = \sup_{x \neq \theta}{\frac{\|Ax\|}{\|x\|}}. $$
Записывая $x \neq \theta$ в виде
$$ x = \|x\|z, $$
где $z$ - вектор единичной длины: $\|z\| = 1$, получаем эквивалентное представление нормы матрицы:
$$ \| A \| = \sup_{\|z\| = 1}{\|Az\|}. $$

Также, из определения следует, что
$$ \|Ax\| \leq \|A\| \|x\|. $$

Будем называть математическую задачу *корректной (по Адамару)*, если:
1. Решение существует;
2. Решение единственно;
3. Решение непрерывно зависит от входных данных.

Рассмотрим СЛАУ
$$ Ax = f,\quad f = \{f_1, f_2, \ldots f_n\} \in E_n,\quad det\, A \neq 0. $$

Можем выразить $x$:
$$ x = A^{-1}f. $$

Пусть правая часть СЛАУ подверглась возмущению $\,\delta f$ и стала равной $\,\tilde f = f + \delta f.$
Тогда решение $\tilde x$ возмущенной системы 
$$ A \tilde x = \tilde f $$
также можно записать через обратную матрицу $A^{-1}$:
$$ \tilde x = A^{-1} \tilde f = A^{-1}f + A^{-1} \delta f = x + \delta x, $$
где
$$ \delta x = A^{-1} \delta f. $$
Отсюда получаем
$$ \| \delta x \| \leq \| A^{-1} \| \| \delta f \|. $$

Последнее неравенство показывает непрерывною зависимость возмущения решения $\delta x$ от правой части $\delta f$:
$$ \|\delta x\| \rightarrow 0 \quad\text{при}\quad \|\delta f\| \rightarrow 0. $$

Исходное уравнение позволяет записать:
$$ \|f\| \leq \|A\|\|x\|. $$
Домножив его на неравенство выше получим:
$$ \|f\| \|\delta x\| \leq \|A\|\|A^{-1}\|\|x\| \|\delta f\|. $$
Пусть $f \neq 0$, тогда неравенство можно переписать в виде:
$$ \frac{\|\delta x\|}{\|x\|} \leq M_A \frac{\|\delta f\|}{\|f\|}, $$
где
$$ M_A = \|A\|\|A^{-1}\|. $$

Число $M_A$ называется *числом обусловленности матрицы $A$*. Она позволяет оценить относительную погрешность решения через относительную погрешность возмущения правой части. Чем больше $M_A$, тем резче реагирует решение на возмущение правой части.

*По хорошему, здесь должен быть пример о нахождении числа обусловленности, но он прост в понимании и сложен для записи. Основная идея такова: ищем нормы матриц  $A$ и $A^{-1}$ и перемножаем их (все просто).
Подробнее: Задача 1 в учебнике Костомарова*

#### Оценка числа обусловленности.
**Утверждение.** Для числа обусловленности матрицы $A$ справедливо неравенство:
$$ M_A \geq \frac{\left|\lambda_{\text{max}}\right|}{\left|\lambda_{\text{min}}\right|}. $$
($\lambda_{\text{min}} \neq 0$, так как $det\, A \neq 0. $)

$\blacktriangle$ 
Пусть $y$ - собственный вектор, отвечающий собственному значению $\lambda_{\text{max}}$:
$$ Ay = \lambda_{\text{max}}y,$$
тогда
$$|\lambda_{\text{max}}|\|y\| = \|Ay\| \leq \|A\|\|y\|, $$
откуда, так как $\|y\| \neq 0$:
$$|\lambda_{\text{max}}| \leq \|A\|, $$

Пусть $z$ - собственный вектор, отвечающий собственному значению $\lambda_{\text{min}}$:
$$ Az = \lambda_{\text{min}}z.$$
Запишем в виде
$$ \frac{1}{\lambda_{\text{min}}} z = A^{-1}z, $$
тогда
$$\frac{1}{|\lambda_{\text{min}}|} \leq \|A^{-1}\|, $$

Перемножая два последних неравенства придем к утверждению. $\blacksquare$

Если матрица симмтеричная ($A = A^*$), то все ее характеристические значения вещественны, причем
$$ \|A\| = |\lambda_{\text{max}}|,\quad \|A^{-1}\| = \frac{1}{|\lambda_{\text{min}}|}, $$
$\blacktriangle$ Пусть $\{e_1,\ldots , e_n\}$ - ОНБ из собственных векторов (такой можно построить только для самосопряженного оператора). Тогда $\forall x$
$$ x = \sum_{i=1}^n{x_ie_i}, \quad \|x\|^2 = \sum_{i=1}^n{x_i^2}, $$
$$ Ax = \sum_{i=1}^n{\lambda_ix_ie_i}, \quad \|Ax\|^2 = \sum_{i=1}^n{\lambda_i^2x_i^2}, $$
Т.о.
$$ \|Ax\|^2 \leq \lambda_{\text{max}}^2\|x\|^2
\quad\Rightarrow\quad
\|A\| \leq |\lambda_{\text{max}}|.$$
Сопоставляя с 
$$ \|A\| \geq |\lambda_{\text{max}}|
\quad\Rightarrow\quad
\|A\| = |\lambda_{\text{max}}|. \blacksquare$$

Поэтому
$$ M_A = \frac{\left|\lambda_{\text{max}}\right|}{\left|\lambda_{\text{min}}\right|}. $$

#### Полная оценка относительной погрешности.

Предположим, вызмущены и правая часть, и коэффициенты:
$$ \tilde A x = \tilde f. $$

**Теорема.** Пусть матрица $A$ имеет обратную и выполнено условие
$$ \| \delta A \| < \frac{1}{\| A^{-1} \|}. $$
Тогда матрица $\tilde A = A + \delta A$ имеет обратную и справедлива оценка погрешности:
$$ \frac{\| \delta x \|}{\| x \|} \leq
\frac{M_A}{1 - M_A \frac{\| \delta A \|}{\| A \|}} 
\left ( \frac{\| \delta A \|}{\| A \|} + \frac{\| \delta f \|}{\| f \|} \right ). $$

$\blacktriangle$ Доказательство в учебнике Самарского. $\blacksquare$
